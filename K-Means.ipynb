# K-Means Clustering 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
univ = pd.read_csv('Universities.csv')
```

```python
univ
```

```python
# Normalize the Data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_univ = scaler.fit_transform(univ.iloc[:,1:])
```

```python
scaled_univ
```

```python
# build clsutering model


cluster_new = KMeans(3, random_state = 42)
cluster_new.fit(scaled_univ)
```

```python
cluster_new.labels_
```

```python
univ['cluster_id'] = cluster_new.labels_
```

```python
univ.sort_values('cluster_id')
```

```python
cluster_new.cluster_centers_
```

```python
univ.groupby('cluster_id').agg(['mean']).reset_index()
```
